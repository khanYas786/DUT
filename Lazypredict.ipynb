```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
vectorizer = HashingVectorizer(n_features=2**10)

tfidf  = TfidfVectorizer()

df_csvLP = pd.read_csv('Text-label-xl-cleaned.csv')

df_csvLP = df_csvLP.dropna()

X = df_csvLP['Text']
train = vectorizer.fit_transform(X)
#train = tfidf.fit_transform(X)
```

```python
train = train.astype('float32')
train = train.toarray()
```

```python
y = df_csvLP['Label']
#X = data.data
#y= data.target

X_train, X_test, y_train, y_test = train_test_split(train, y,test_size=.25,random_state =123)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
```

```python
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
```
